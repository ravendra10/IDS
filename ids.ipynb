```python
import random
sample_size=random.randint(10,986)
print(sample_size)
```

```python
sample_indices = set();
for i in range(0,sample_size):
    sample_indices.add(random.randint(0,985))
print(sample_indices)
```

```python
sample = []
#df['sq__ft'].fillna(df['sq__ft'].mean(), inplace=True)
#print(df['sq__ft'].isnull().sum())
print(df['sq__ft'])
for i in sample_indices:
    sample.append(df['sq__ft'][i]);
#print(sample)
samplemean = np.mean(sample)
samplestd = np.std(sample)

popmean = np.mean(df['sq__ft']);

print(samplemean)
print(popmean)

```

```python
z_star=(samplemean-popmean)/samplestd
print(z_star)
p_value=2*norm.cdf(z_star)
print(p_value)
```

```python
actual_z = abs(norm.ppf(0.05/2))

hypo_z = z_star;

if (hypo_z<= actual_z or hypo_z >= -(actual_z)):
    print(True)
else:
    print(False)
```
